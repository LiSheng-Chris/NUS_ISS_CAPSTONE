```python
# dilation
img = cv2.imread("./data/masks/mask_ZT199_1_A_3_11.png",cv2.IMREAD_COLOR)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

kernel = np.ones((250,250),np.uint8)
new_image =cv2.morphologyEx(np.uint8(img),cv2.MORPH_CLOSE,kernel)
print ('new_imag shape: ',new_image.shape)


fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(10, 10))

ax0.imshow(img)
ax0.set_title("orignal RGB image")
ax0.axis('off')
ax1.imshow(new_image, cmap='hsv')
ax1.set_title("dilation image")
ax1.axis('off')

fig.tight_layout()
```

```python
#Haematoxylin-Eosin colour space
import cv2
import shutil 

np.random.seed(42)
mask_dir= './data/masks/'
mask_aug_dir='./data/masks_morphology/'

def hed_aug(image_path):
  plt.axis('off')
  kernel = np.ones((250,250),np.uint8)
  
  image_name = os.path.splitext(os.path.basename(image_path))[0]
  print('image_name : ',image_name)

  mask_name='mask_'+image_name+'.png'
  mask_aug_name='mask_'+image_name+'.png'
  mask_aug_path= os.path.join(mask_aug_dir,mask_aug_name)
  print('mask_aug_name : ',mask_aug_path)
    
  # generate mask - morphology
  if os.path.isfile(mask_aug_path):
    print (mask_aug_path, 'existed , skipping')
  else:
    mask_path=mask_dir+mask_name
    mask= cv2.imread(mask_path, cv2.IMREAD_COLOR)
    mask = cv2.cvtColor(mask,cv2.COLOR_BGR2RGB)
    print('mask shape :',mask.shape)

    new_mask =cv2.morphologyEx(np.uint8(mask),cv2.MORPH_CLOSE,kernel)
    print ('new_imag shape: ',new_mask.shape)
    imgplot = plt.imshow(new_mask)
    #plt.show()

    print('mask_aug_path : ',mask_aug_path)
    plt.savefig(mask_aug_path, transparent = True, bbox_inches = 'tight', pad_inches = 0)
```

```python
import os
rootdir = './data/images/'

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        hed_aug(os.path.join(subdir, file))
```
